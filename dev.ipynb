```python
x= np.ones(2)
y= np.ones(2)

# Creamos un conjunto de datos en formato tensor
data_set = TensorDataset(torch.from_numpy(x.astype('float32')), 
                          torch.from_numpy(y.astype('float32')))

# Data loader de entrenamiento
data_loader = DataLoader(data_set, shuffle=True, batch_size=32)
```
